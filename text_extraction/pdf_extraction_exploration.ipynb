# PyMuPDF
https://pymupdf.readthedocs.io/en/latest/ 

```python
import pymupdf

mercedes_rep = pymupdf.open("mercedes-2024.pdf")
axa_rep = pymupdf.open("AXA-2024.pdf")
```

```python
# extract text from page 132 of the AXA report
axa_text = axa_rep[132].get_text("text")
print(axa_text)
```

--> PyMuPDF not capable tho extract the images with labeling

```python
# extract text from page 172 of the AXA report
axa_text = axa_rep[172].get_text("text")
print(axa_text)
```

--> extracted the content, but not in such a way that it is easy further usable

```python
### use the option specifically for tables ###

from pprint import pprint

page = axa_rep[172] # get the relevant page of the document
tabs = page.find_tables() # locate and extract any tables on page
print(f"{len(tabs.tables)} found on {page}") # display number of found tables

if tabs.tables:  # at least one table found?
   pprint(tabs[0].extract())  # print content of first table
```

--> could not extract the table even in the specific function for tables

```python
# extract text from page 129 of the Mercedes report
mercedes_text = mercedes_rep[129].get_text("text")
print(mercedes_text)
```

--> extracted everything from the table

```python
# extract text from page 167 of the Mercedes report
mercedes_text = mercedes_rep[167].get_text("text")
print(mercedes_text)
```

# Docling
https://github.com/docling-project/docling 
https://docling-project.github.io/docling/

```python
from docling.document_converter import DocumentConverter
```

```python
doc = pymupdf.open("AXA-2024.pdf") # open the document
doc.select([129, 172]) # select only relevant pages of the document
doc.save("AXA_only_pages.pdf") # save the document
```

```python
# Convert the selected page to markdown
source = "AXA_only_pages.pdf"
converter = DocumentConverter()
result = converter.convert(source)
print(result.document.export_to_markdown())  # output: "## Docling Technical Report[...]"
```

```python
doc = pymupdf.open("Mercedes-2024.pdf") # open the document
doc.select([129, 167]) # select only relevant pages of the document
doc.save("Mercedes_pages_only.pdf") # save the document

# Convert the selected page to markdown
source = "Mercedes_pages_only.pdf"
converter = DocumentConverter()
result = converter.convert(source)
print(result.document.export_to_markdown())  # output: "## Docling Technical Report[...]"
```
